```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
data = pd.read_csv('Desktop/theoria_apofasewn_2/diabetes.csv')
data.head()
```

```python
display(data.info()) #8 einai ta xaraktiristika kai to outcome
```

```python
# erwtima 1
print(data.shape) #to prwto apo to shape einai oi grammes kai ta xaraktiristika -to 9 einai to outcome dn to metraw
print(data[data['Outcome']==1].count()) #268 stin class 1
print(data[data['Outcome']==0].count()) #500 sstin class 0
```

```python
target = data['Outcome']
del data['Outcome']
print(target.head())
print(data.head())
```

```python
data['Age'].max()
```

```python
max_abs_scaler = preprocessing.MaxAbsScaler()
data_scaled = max_abs_scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled)
data_scaled
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
scores = cross_val_score(clf, data_scaled,target,cv=5) # ama valw parametro scoring = precision pernw ta scores gia akriveia
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,make_scorer,recall_score

clf = GaussianNB()
# conf_scorer = make_scorer(confusion_matrix)
specificity = make_scorer(recall_score,pos_label=0) # auto einai to specificity 
# NA TO TSEKARW ME TP,TN,FP,FN ΑΝ ΕΙΝΑΙ SWSTO
scores = cross_validate(clf,data_scaled,target,cv=5,scoring=specificity)
print(scores.keys())

print(f"Specificity: {scores['test_score']}") 
print(f"Mean specificity: {np.mean(scores['test_score'])}")
```

```python
#AUTO EINAI TO ERWTIMA 3 GIA TON BAYES MONO
from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import make_scorer,recall_score
import math

clf = GaussianNB()
specificity = make_scorer(recall_score,pos_label=0) # auto einai to specificity 
scoring = {'sensitivity':'recall','specificity':specificity}
scores = cross_validate(clf,data_scaled,target,cv=5,scoring=scoring)
print(scores.keys())
print(f"sensitivity: {scores['test_sensitivity']}")
print(f"Mean sensitivity: {np.mean(scores['test_sensitivity'])}")
print(f"specificity: {scores['test_specificity']}")
print(f"Mean specificity: {np.mean(scores['test_specificity'])}")
print(f"Geometric Mean: {math.sqrt(np.mean(scores['test_sensitivity'])*np.mean(scores['test_specificity']))}")
```

```python
#ERWTIMA 4 - SVM
from sklearn import svm
for c in range(1,200,5):
    print(f"C={c}")
    clf = svm.SVC(kernel='linear',C=c)
    scores = cross_validate(clf,data_scaled,target,cv=5,scoring=scoring)
    print(f"Mean sensitivity: {np.mean(scores['test_sensitivity'])}")
    print(f"Mean specificity: {np.mean(scores['test_specificity'])}")
    print(f"Geometric Mean: {math.sqrt(np.mean(scores['test_sensitivity'])*np.mean(scores['test_specificity']))}")
    
# to veltisto C vgainei 46         
    
```

```python
# ERWTIMA 4
for gamma in [x*0.5 for x in range(1,21)]:
    print(f"gamma={gamma}")
    clf = svm.SVC(kernel='rbf',C=46,gamma=gamma)
    scores = cross_validate(clf,data_scaled,target,cv=5,scoring=scoring)
    print(f"Mean sensitivity: {np.mean(scores['test_sensitivity'])}")
    print(f"Mean specificity: {np.mean(scores['test_specificity'])}")
    print(f"Geometric Mean: {math.sqrt(np.mean(scores['test_sensitivity'])*np.mean(scores['test_specificity']))}")
# gamma=2 einai i kaliteri epiologi mallon
```

```python
# ERWTIMA 4
from sklearn.neighbors import KNeighborsClassifier
for k in range(3,11):
    print(f"k={k}")
    clf = KNeighborsClassifier(n_neighbors=k)
    scores = cross_validate(clf,data_scaled,target,cv=5,scoring=scoring)
    print(f"Mean sensitivity: {np.mean(scores['test_sensitivity'])}")
    print(f"Mean specificity: {np.mean(scores['test_specificity'])}")
    print(f"Geometric Mean: {math.sqrt(np.mean(scores['test_sensitivity'])*np.mean(scores['test_specificity']))}")
# k=5 einai i kaliteri epilogi edw
```

```python
# to apo panw me xrisi tis cross_validate
from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
scoring = ['precision','recall','accuracy']
scores = cross_validate(clf, data_scaled, target, cv=5,scoring=scoring)
print(scores.keys())
print(f"precision: {scores['test_precision']}")
print(f"recal: {scores['test_recall']}")
print(f"default accuracy: {scores['test_accuracy']}")
```
